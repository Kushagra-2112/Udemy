```python
arr = np.array([
    [
        [1,2,3,4],
        [4,5,6,7]
    ],
    [
        [2,3,4,1],
        [2,4,1,1]
    ]
])

print(arr)
type(arr)
print(arr.ndim)
print(arr.astype("float"))


for i in arr:
    for j in i:
        for k in j:
            print(k)
```
