```python
class VendingMachine:

    def __init__(self):
        self.__balance = 0
        self.items = {
            "Soda" : { "Price": 15.00, "Stock": 5},
            "Chips" : { "Price": 10.00, "Stock": 5},
            "Cold-drinks": { "Price": 20, "Stock":5},
            "Candy": { "Price":20, "Stock":10}
        }

    def show_items(self):
        print("Availiable Items are: ")

        for item,info in self.items.items():
            print(f"{item} : {info['Price']} and {info['Stock']}")

    def deposit_money(self,amount):
        if amount > 0:
            self.__balance += amount
            print("Amount depoisited...done")
        else:
            print("Please Enter valid amount")

    def purchase_item(self,item_name):
        item = self.items.get(item_name)

        if item:
            if item['Stock'] > 0:
                if self.__balance >= item['Price']:
                    print("Purchase Successul...done")
                    self.__balance -=  item['Price']
                    item['Stock'] -= 1
                else:
                    print("Insufficient Funds...")

            else:
                print("Out of Stock...")

        else:
            print("Item not found")

    def check_balance(self):
        print("Balance is: ", self.__balance)

```

```python
Customer1 = VendingMachine()
```

```python
Customer1.show_items()
```

```python
Customer1.deposit_money(100)
```

```python
#Customer1.__balance() --> private no one can see
```

```python
Customer1.purchase_item("Chips")
```

```python
Customer1.check_balance()
```

```python
class Fraction:
    def __init__(self,num,den):
        self.num = num
        self.den = den

    def __str__(self): #automatically runs print statement is there
        return f"{self.num}/{self.den}"
    
    def __add__(self,other):
        new_num = (self.num * other.den) + (other.num * self.den)
        new_den = self.den * other.den
        return f"{new_num}/{new_den}"
    
```

```python
obj1 = Fraction(2,3)
```

```python
obj2 = Fraction(3,4)
```

```python
obj1 + obj2 
```
