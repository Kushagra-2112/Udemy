```python
#Idendity operators
#is: Returns True if both variables point to the same object.
#is not: Returns True if the variables point to different objects.

# Two separate lists with identical contents
list_a = [1, 2, 3]
list_b = [1, 2, 3]
list_c = list_a  # c points to the exact same object as a

print(list_a == list_b) # True  -> Because their VALUES are equal
print(list_a is list_b) # False -> Because they are distinct objects in memory
print(list_a is list_c) # True  -> Because they share the exact same memory address
```

```python
#Bitwise Operators (&, |, ^, ~, <<, >>)

# AND	Sets bit to 1 if both bits are 1.
# OR	Sets bit to 1 if at least one bit is 1.
# XOR	Sets bit to 1 if only one bit is 1.
# NOT	Inverts all the bits (flips 1 to 0 and 0 to 1).
# Left Shift	Shifts bits to the left, filling in with 0s.
# Right Shift	Shifts bits to the right, dropping rightmost bits.
```
