```python
#Join

#union
se4 = {1,2,3,4,5}
se5 = {5,6,3,7,8}

se6 = se4.union(se5)
print(se6)

se7 = se4.intersection(se5)
print(se7)

se8 = se4.difference(se5)
print(se8)

se9 = se4^se5 #uncommon elements make a set symmetric difference
print(se9)


```

```python
#map, filter work only on iterables

#map(fun,iterable) #two arguments and thenc convert output in form of list as it will give in obj

numbers = [2,3,4,5]
sq = list(map(lambda x:x*x,numbers))
print(sq)
```

```python
#filter(function,iterable) expression true create a new list of them
list11 = [2,4,5,6]
div = list(filter(lambda x:x%2==0, list11))
print(div)
```

```python
def infinite_counter():
    num = 1
    while True:
        yield num
        num += 1  # Agli baar ke liye value badha di

# Generator object banaya
counter = infinite_counter()

print(next(counter))  # Output: 1
print(next(counter))  # Output: 2
print(next(counter))
```
