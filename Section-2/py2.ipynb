```python
#sets

se = {1,2,3,4,4} #does not allow dduplicates count duplicates only once
print(se)

#se[0] unordered not fix position but mutable
se.add(5)
print(se)

se2 = {33,44}
se.update(se2)
print(se)
```

```python
se.remove(33)
se
```

```python
se.discard(33) #if also not there it will not remove it
```

```python
se.pop() #as unordered no index randomly remmove
```

```python
#Join

#union
se4 = {1,2,3,4,5}
se5 = {5,6,3,7,8}

se6 = se4.union(se5)
print(se6)

se7 = se4.intersection(se5)
print(se7)

se8 = se4.difference(se5)
print(se8)

se9 = se4^se5 #uncommon elements make a set symmetric difference
print(se9)


```

```python
se10 = {1,2,3,4}
se10.issubset(se4)
```
